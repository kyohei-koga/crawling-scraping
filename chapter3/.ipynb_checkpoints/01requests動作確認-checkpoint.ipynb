```python
import requests
```

```python
r = requests.get('https://sample.scraping-book.com/dp') #get関数でwebページを取得できる
```

```python
type(r) #getの戻り値はresponse型
```

```python
r.status_code #status_code属性でHTTPステータスコードを取得できる
```

```python
r.headers['content-type'] #headers属性でHTTPヘッダーの辞書を取得できる
```

```python
r.encoding # encoding属性でHTTPヘッダーから得られたエンコーディングを取得できる
```

```python
r.text #text属性でstr型にデコードしたレスポンスボディを取得できる。
```

```python
r.content # content属性でbytes型のレスポンスボディを取得できる
```

```python
#Livedoor 天気webサービスで東京の天気をJSON形式で取得する
r = requests.get('http://weather.livedoor.com/forecast/webservice/json/v1?city=130010')
```

```python
r.status_code
```

```python
r.json()
```

```python
#POSTメソッドで送信。キワード引数dataにdictを指定するとHTMLフォーム形式で送信される
r = requests.post('http://httpbin.org/post',data={'key1':'value1'})
```

```python
#リクエストを追加するHTTPヘッダーをキーワード引数headersにdictで指定する。
r = requests.get('http://httpbin.org/get',
                 headers={'user-agent':'my-crawler/1.0 (+foo@example.com)'})
```

```python
#Basic認証のユーザー名とパスワードの組をキーワード引数authで指定する。
r = requests.get('http://api.github.com/user',
                 auth=('githubのユーザーID','githubのパスワード'))
```

```python
#URLのパラメータはキーワード引数paramsで指定することも可能
r = requests.get('http://httpbin.org/get',params={'key1':'value1'})
```

```python
s = requests.Session()
```

```python
# HTTPヘッダーを複数のリクエストで使いまわす。
s.headers.update({'user-agent':'my-crawler/1.0 (+foo@example.com)' })
```

```python
#Sessionオブジェクトにはget(),post()などのメソッドがあり、requests.get(),requests.post()などと同様に使える。
r = s.get('https://gihyo.jp/')
```

```python
r.status_code
```

```python
r = s.get('https://gihyo.jp/dp')
```

```python
r.status_code
```
