```python
import lxml.html
from pymongo import MongoClient
```

```python
#HTMLファイルを読み込みgetroot()メソッドでHtmlElementsオブジェクトを得る。
tree = lxml.html.parse('/home/kyohei/index.html')
html = tree.getroot()
```

```python
client = MongoClient('localhost',27017)
```

```python
db = client.scraping #scrapingデータベースを取得する
collection = db.links #linksオブジェクトを取得する
```

```python
#このスクリプトを何回実行しても同じ結果になるようにするため、コレクションのドキュメントを全て削除する。
collection.delete_many({})
```

```python
#cssselect()メッソドでa要素のリストを取得して、個々のa要素に対して処理を行う。
for a in html.cssselect('a'):
    #href属性とリンクとのテキストを取得して保存する。
    collection.insert_one({
        'url':a.get('href'),
        'title':a.text,
    })
```

```python
#コレクションの全てのドキュメント_idの順にソートして取得する。
for link in collection.find().sort('_id'):
    print(link['_id'],link['url'],link['title'])
```
