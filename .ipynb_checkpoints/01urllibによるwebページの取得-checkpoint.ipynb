# urllibによるwebページの取得

```python
from urllib.request import urlopen
```

```python
#urlopen()関数はHTTPResponse型のオブジェクトを返す。
#このオブジェクトはファイルオブジェクトであり、組み込み関数open()の戻り値であるファイルオブジェクトと同じように扱える。
f = urlopen('https://sample.scraping-book.com/dp')
```

```python
type(f)
```

```python
#read()メソッドではHTTPレスポンスのボディ(bytes型)を取得できる。
#HTTPコネクションは自動的に閉じられるので、明示的にclose()を呼び出す必要は無い。
f.read()
```

```python
#ステータスコードを取得
f.status
```

```python
#HTTPヘッダーの値を取得する。
f.getheader('Content-Type') 
```
