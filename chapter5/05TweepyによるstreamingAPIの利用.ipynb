```python
import tweepy
import pandas as pd
```

```python
api_info = pd.read_csv('api.csv',header=None,names=['name', 'value'])
```

```python
CONSUMER_KEY = api_info.loc[0, 'value']
CONSUMER_SECRET = api_info.loc[1, 'value']
ACCESS_TOKEN = api_info.loc[2, 'value']
ACCESS_TOKEN_SECRET = api_info.loc[3, 'value']
```

```python
# 認証情報を設定する。
auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)
```

```python
class MyStreamLinstener(tweepy.StreamListener):
    """
    Streaming APIで取得したツイートを処理するためのクラス。
    """
    
    def on_status(self, status):
        """
        ツイートを受信した時に呼び出されるメソッド。引数はツイートを表すStatusオブジェクト。
        """
        print('@' + status.author.screen_name, status.text)
```

```python
# 認証情報とStreamLinstenerを指定してStreamオブジェクトを取得する。
stream = tweepy.Stream(auth, MyStreamLinstener())
# 公開されているツイートをサンプリングしたストリームを受信する。
# キワード引数languagesで、日本語のツイートのみに絞り込む。
stream.sample(languages=['ja'])
```
